```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df.set_index("Date",inplace=True)
```

# Exploratory Analysis & Visualization

# Stock Price Trend Analysis

```python
#Line Chart (Closing Price Over Time)
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Close', data=df)
plt.title("Closing Price Over Time")
plt.show()
```

```python
#Line Chart (Adj_Close Price Over Time)
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date', y='Adj_Close', data=df)
plt.title("Adj_Close Price Over Time")
plt.show()
```

```python
#Bar Chart (Trading Volume Over Time)
plt.figure(figsize=(10,5))
sns.barplot(x='Year', y='Volume', data=df,ci=None,color="grey")
plt.xticks(rotation=45)
plt.title("Trading Volume Over Time")
plt.show()
```

```python
#Pairplot (Relationships Between Prices)
sns.pairplot(df[["Open", "High", "Low", "Close"]])
plt.show()
```

```python
# Heatmap (Correlation Between Variables)
sns.heatmap(df[["Open", "High", "Low", "Close", "Volume"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# KDE Plot (Price Density)
sns.kdeplot(data=df, x="Close", fill=True)
plt.title("Density Plot of Closing Prices")
plt.show()
```
